```python
import pandas as pd
import plotly.express as px
import hvplot.pandas
from sklearn.cluster import KMeans
```

```python
file_to_load=pd.read_csv('merged_data.csv')
```

```python
data_df = pd.DataFrame(file_to_load)
```

```python
data_df = data_df[data_df['price'] < 4000000]  
```

```python
data_df.head(10)
```

```python
fig1 = px.scatter_3d(data_df, x="house_age", y="price",z="city_rank",
                     color="bedrooms",size="sqft_lot")
fig1.update_layout(title="5 Features Representation")
fig1.show()
```

```python
inertia = []
k = list(range(1, 11))
```

```python
# Looking for the best K
for i in k:
    km = KMeans(n_clusters=i, random_state=0)
    km.fit(data_df)
    inertia.append(km.inertia_)
```

```python
# Define a DataFrame to plot the Elbow Curve using hvPlot
elbow_data = {"k": k, "inertia": inertia}
df_elbow = pd.DataFrame(elbow_data)
df_elbow.hvplot.line(x="k", y="inertia", title="Elbow Curve", xticks=k)
```

```python
def get_clusters(k, data):
   # Create a copy of the DataFrame
   data = data_df.copy()
 
   # Initialize the K-Means model
   model = KMeans(n_clusters=k, random_state=0)

   # Fit the model
   model.fit(data)

   # Predict clusters
   predictions = model.predict(data)

   # Create return DataFrame with predicted clusters
   data["class"] = model.labels_

   return data_df
```

```python
three_clusters = get_clusters(3, data_df)
three_clusters.head()
```

```python
four_clusters = get_clusters(4, data_df)
four_clusters.head()
```

```python
# Plot the 3D-scatter with x="Annual Income", y="Spending Score (1-100)" and z="Age"
fig = px.scatter_3d(
    three_clusters,
    x="price",
    y="bedrooms",
    z="bathrooms",
    width=800,
)
fig.update_layout(legend=dict(x=0, y=1))
fig.show()
```
