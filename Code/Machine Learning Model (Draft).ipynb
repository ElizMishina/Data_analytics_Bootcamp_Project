```python
# Calculate Pearsons Correlation coefficient
corr_df = cleaned_df.corr(method="pearson")
corr_price = corr_df['price'].values
corr_df = corr_df.drop(index="price")
corr_df = corr_df.sort_values(by=['price'],ascending=False)
corr_df.plot(kind ="bar",y="price")
```

```python
# Select features that have correlation greater that 0.5
sel_feature_pearsons = corr_df[corr_df["price"]>0.5].index.values.tolist()
sel_feature_pearsons
```

```python
# Filter for selected features only
corr_test = cleaned_df.filter(['sqft_living','grade','sqft_above', 'bathrooms'], axis=1)
corr_test
```

```python
# Create independant & dependant variables
X = corr_test
y = cleaned_df['price']
```

```python
# Split the data into training/testing sets
X_train = X[:80]
X_test = X[80:]
y_train = y[:80]
y_test = y[80:]

#Transforming y_train to log
y_train = np.log(y_train)

# Creating a StandardScaler instance.
scaler = StandardScaler()
# Fitting the Standard Scaler with the training data.
X_scaler = scaler.fit(X_train)
# Scaling the data.
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Create linear regression object
regr = LinearRegression()
```

```python
# Train the model using the training sets
regr.fit(X_train, y_train)
```

```python
# Make predictions using the testing set
y_pred = regr.predict(X_test)
```

```python
# The coefficients
print('Coefficients: \n', regr.coef_)
# The mean squared error
print('Mean squared error: %.2f'
      % mean_squared_error(y_test, y_pred))
# The coefficient of determination: 1 is perfect prediction
print('Coefficient of determination: %.2f'
      % r2_score(y_test, y_pred))
```

```python
# Create a regressor model.
rf_model = RandomForestRegressor(n_estimators=500) 
# Fitting the model
rf_model = rf_model.fit(X_train, y_train)
# Making predictions using the testing data.
y_pred_rf = rf_model.predict(X_test)
y_pred_rf = np.exp(y_pred_rf)
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred_rf))
print('R_Squared:', metrics.r2_score(y_test, y_pred_rf))
```
